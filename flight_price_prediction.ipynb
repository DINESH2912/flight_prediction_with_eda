```python
data['Dep_Time']=data['Dep_Time'].apply(lambda x : x.split(' ')[0])
data['Dep_hour']=data['Dep_Time'].str.split(':').str[0]
data['Dep_min']=data['Dep_Time'].str.split(':').str[1]
```

```python
data['Hour']=data['Duration'].str.split('h').str[0]
data.drop(6474,axis=0,inplace=True)
data.drop(2660,axis=0,inplace=True)
data[data['Hour']=='5m']
```

```python
data['Hour']=data['Hour'].astype('int')
```

```python
data['Hour']=data['Hour']*60
data
```

```python
data=data.drop('Duration',axis=1)
```

```python
data=data.drop('Route',axis=1)
data
```

```python
data['Total_Stops'].value_counts()
```

```python
data['Total_Stops']=data['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
data.info()
```

```python
data.drop('Additional_Info',axis=1)
```

```python
data['Airline']=data['Airline'].reshape(-1,1)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

onehot_encoder = OneHotEncoder(sparse=False)

# One-hot encode 'Airline' and add the result to the DataFrame
airline_encoded = onehot_encoder.fit_transform(data['Airline'].values.reshape(-1, 1))
airline_columns = onehot_encoder.get_feature_names_out(['Airline'])
data[airline_columns] = pd.DataFrame(airline_encoded, columns=airline_columns)

# One-hot encode 'Destination' and add the result to the DataFrame
destination_encoded = onehot_encoder.fit_transform(data['Destination'].values.reshape(-1, 1))
destination_columns = onehot_encoder.get_feature_names_out(['Destination'])
data[destination_columns] = pd.DataFrame(destination_encoded, columns=destination_columns)

# One-hot encode 'Source' and add the result to the DataFrame
source_encoded = onehot_encoder.fit_transform(data['Source'].values.reshape(-1, 1))
source_columns = onehot_encoder.get_feature_names_out(['Source'])
data[source_columns] = pd.DataFrame(source_encoded, columns=source_columns)

# Drop the original categorical columns
data.drop(['Airline', 'Destination', 'Source'], axis=1, inplace=True)

```

```python
data
```

```python

data.dropna(subset=['Price'], inplace=True)
data
```

```python
data.dropna(subset=['Total_Stops'], inplace=True)
data
```

```python
y=data['Price']
y.isnull().sum()
```

```python
x=data.drop(columns=['Price'])
x=data.drop(columns=['Additional_Info'])
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error:", mse)
```
